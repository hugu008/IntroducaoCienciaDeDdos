```python
# Importar a biblioteca
import pandas as pd
```

```python
#carregar arquivo
df = pd.read_csv('vendas_de_ingressos.csv')
```

```python
#1- Exibir a quantidade de linhas e colunas do data frame.
df.shape
```

```python
#2- Exibir as 7 primeiras linhas do data frame.
df.head(7)
```

```python
#3- Exibir as 10 últimas linhas do data frame.

df.tail(10)
```

```python
#4- Renomear as colunas de forma permanente:
# Age = Idade
# Ticket_Price = Preco_do_Ingresso
# Movie_Genre = Genero
# Seat_Type = Tipo_de_Assento
# Number_of_Person = Numero_de_Pessoas
# Purchase_Again = Compraria_Novamente

df.rename(columns = {
    'Age': 'Idade',
    'Ticket_Price': 'Preco_do_Ingresso',
    'Movie_Genre': 'Genero',
    'Seat_Type': 'Tipo_de_Assento',
    'Number_of_Person': 'Numero_de_Pessoas',
    'Purchase_Again': 'Compraria_Novamente'
}, inplace = True)

df.head(0)
```

```python
#5- Resumo estatístico dos dados numéricos.

df.describe()
```

```python
#6- Filtrar os dados por gênero.

df[['Genero']]
```

```python
#7- Exibir os dados das linhas 12 a 20 (inclusive).

df.iloc[11:20]
```

```python
#8- Exibir os dados das linhas 35, 38, 42 e 50.

df.iloc[[34, 37, 41, 49]]
```

```python
#9- Exibir os dados em que o tipo de assento é Standard.

df.loc[df["Tipo_de_Assento"] == "Standard"]
```

```python
#10- Exibir os dados em que as idades dos consumidores está entre 20 (inclusive) e 45 anos (inclusive).

df.loc[(df["Idade"] >= 20) & (df["Idade"] <= 45)]
```

```python
#11- Criar um novo data frame em que o gênero dos filmes assistidos é Comedy.

df_comedy = df.loc[df["Genero"] == "Comedy"]
df_comedy
```

```python
#12- Criar um novo data frame exibindo os registros de Preço do Ingresso, Gênero e Tipo de Assento nos clientes que comprariam novamente.

df_novamente = df.loc[df["Compraria_Novamente"] == "Yes"][["Preco_do_Ingresso", "Genero", "Tipo_de_Assento"]]
df_novamente
```

```python
#13- Criar uma nova series filtrando os dados pelo Tipo de Assento.

series_assento = df["Tipo_de_Assento"]
print(series_assento)
```

```python
#14- Excluir a coluna Ticket_ID de forma permanente.

df.drop(columns=["Ticket_ID"], inplace=True)

df.head(0)
```
